```python
import os
import cv2
import numpy as np
import pandas as pd

df = pd.read_csv("/content/drive/MyDrive/Signature_Project/pairs.csv")

ORG_DIR = "/content/drive/MyDrive/Signature_Project/processed/full_org"
FORG_DIR = "/content/drive/MyDrive/Signature_Project/processed/full_forg"

def get_image_path(filename):
    if filename.startswith("original"):
        return os.path.join(ORG_DIR, filename)
    else:
        return os.path.join(FORG_DIR, filename)

X1 = []
X2 = []
y = []

for _, row in df.iterrows():

    img1 = cv2.imread(get_image_path(row["image1"]), 0) / 255.0
    img2 = cv2.imread(get_image_path(row["image2"]), 0) / 255.0

    X1.append(img1)
    X2.append(img2)
    y.append(row["label"])

X1 = np.array(X1)
X2 = np.array(X2)
y = np.array(y)

X1 = X1.reshape(-1,224,224,1)
X2 = X2.reshape(-1,224,224,1)

print(X1.shape)
print(X2.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split

X1_train, X1_test, X2_train, X2_test, y_train, y_test = train_test_split(
    X1,
    X2,
    y,
    test_size=0.2,
    random_state=42
)

print(X1_train.shape)
print(X1_test.shape)
```

```python
from tensorflow.keras.layers import (
    Input,
    Conv2D,
    MaxPooling2D,
    Flatten,
    Dense
)
from tensorflow.keras.models import Model

input_layer = Input(shape=(224,224,1))

x = Conv2D(
    filters=32,
    kernel_size=(3,3),
    activation="relu"
)(input_layer)

x = MaxPooling2D(pool_size=(2,2))(x)

x = Conv2D(
    filters=64,
    kernel_size=(3,3),
    activation="relu"
)(x)

x = MaxPooling2D(pool_size=(2,2))(x)

x = Flatten()(x)

x = Dense(128, activation="relu")(x)

embedding_network = Model(
    inputs=input_layer,
    outputs=x
)

embedding_network.summary()
```

```python
input_a = Input(shape=(224,224,1))
input_b = Input(shape=(224,224,1))

embedding_a = embedding_network(input_a)
embedding_b = embedding_network(input_b)

print(embedding_a)
print(embedding_b)
```

```python
from tensorflow.keras.layers import Lambda
import tensorflow as tf

distance = Lambda(
    lambda tensors: tf.abs(tensors[0] - tensors[1])
)([embedding_a, embedding_b])

print(distance)
```

```python
prediction = Dense(
    1,
    activation="sigmoid"
)(distance)

print(prediction)
```

```python
siamese_model = Model(
    inputs=[input_a, input_b],
    outputs=prediction
)

siamese_model.summary()
```

```python
siamese_model.compile(
    optimizer="adam",
    loss="binary_crossentropy",
    metrics=["accuracy"]
)
```

```python
history = siamese_model.fit(
    [X1_train, X2_train],
    y_train,
    validation_data=([X1_test, X2_test], y_test),
    epochs=5,
    batch_size=32
)
```

```python
loss, accuracy = siamese_model.evaluate(
    [X1_test, X2_test],
    y_test
)

print("Test Loss:", loss)
print("Test Accuracy:", accuracy)
```

```python
import os
import cv2
import numpy as np

ORG_PATH = "/content/drive/MyDrive/Signature_Project/cedar/signatures/full_org"
FORG_PATH = "/content/drive/MyDrive/Signature_Project/cedar/signatures/full_forg"

def predict_signature(img1_name, img2_name):

    # Decide folder for image 1
    if img1_name.startswith("original"):
        img1_path = os.path.join(ORG_PATH, img1_name)
    else:
        img1_path = os.path.join(FORG_PATH, img1_name)

    # Decide folder for image 2
    if img2_name.startswith("original"):
        img2_path = os.path.join(ORG_PATH, img2_name)
    else:
        img2_path = os.path.join(FORG_PATH, img2_name)

    # Read images
    img1 = cv2.imread(img1_path, cv2.IMREAD_GRAYSCALE)
    img2 = cv2.imread(img2_path, cv2.IMREAD_GRAYSCALE)

    # Resize
    img1 = cv2.resize(img1, (224, 224))
    img2 = cv2.resize(img2, (224, 224))

    # Normalize
    img1 = img1 / 255.0
    img2 = img2 / 255.0

    # Reshape
    img1 = img1.reshape(1, 224, 224, 1)
    img2 = img2.reshape(1, 224, 224, 1)

    # Predict
    prediction = siamese_model.predict([img1, img2], verbose=0)

    score = prediction[0][0]

    print("\nPrediction Score:", score)

    if score > 0.5:
        print("✅ Genuine Signature")
        print(f"Confidence: {score*100:.2f}%")
    else:
        print("❌ Forgery")
        print(f"Confidence: {(1-score)*100:.2f}%")
```
